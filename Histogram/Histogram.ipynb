```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data set sportsdata.xls

```python
data = pd.read_excel('sportsdata.xls', usecols=['Age'])
data
```

```python
data.mean() # calculate mean value
```

```python
plt.hist(data['Age'],bins=range(0,55,5),log=True)
plt.plot()
plt.show()
```

```python
fig, ax = plt.subplots()
data.plot.kde(ax=ax, legend=False, title='Age of Player')
data.hist(ax=ax,grid=False,density=True)
plt.xlabel('Age of Players')
plt.ylabel('% of Players')
plt.title('Age Distribution')
plt.plot()
plt.show()
```

```python
sns.set_style('darkgrid')
sns.distplot(data)
```

### Data set OnlineRetail.csv

```python
# read data set 'OnlineRetail.csv'

df = pd.read_csv('OnlineRetail.csv',encoding='latin')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop value missing in 2 column

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

There are 2 columns missing data: Description, CustomerID. 
Remedy: Remove the missing value

```python
df.isnull().sum()
```

```python
df.info()
```

```python
print('Quantity mean: ', df['Quantity'].mean())
print('Quantity median: ', df['Quantity'].median())
```

```python
sns.distplot(df['Quantity'])
line1 = plt.axvline(df['Quantity'].mean(), color='r', linestyle='--', label='Mean')
line2 = plt.axvline(df['Quantity'].median(), color='b', linestyle='--', label='Median')
plt.legend(handles=[line1,line2],loc=1)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Count')
```
